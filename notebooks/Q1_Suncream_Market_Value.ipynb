## Q1. ê²¨ìš¸ ì‹œì¥ì€ íˆ¬ì ê°€ì¹˜ê°€ ìˆëŠ”ê°€? ğŸ¤” 
**ê°€ì„¤:** ê²¨ìš¸ì—ë„ ì„ í¬ë¦¼ ìˆ˜ìš”ê°€ ì¡´ì¬í•˜ë©°, ì—¬ë¦„ ëŒ€ë¹„ ì¼ì • ë¹„ìœ¨ ì´ìƒì„ ì°¨ì§€í•œë‹¤.

**ê²€ì¦ ë°©ë²•:**
- ê³„ì ˆë³„ ê²€ìƒ‰ëŸ‰ ë¹„êµ
- í†µê³„ì  ìœ ì˜ì„± ê²€ì •
- ì—°ë„ë³„ íŠ¸ë Œë“œ ë¶„ì„

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
# í•œê¸€ í°íŠ¸ ì„¤ì •
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
plt.rcParams['axes.unicode_minus'] = False
```

```python
df_trend = pd.read_csv("../data/01_ì„ í¬ë¦¼_ì›”ë³„_íŠ¸ë Œë“œ.csv")
df_trend['date'] = pd.to_datetime(df_trend['date'])
df_trend['month'] = df_trend['date'].dt.month
df_trend['year'] = df_trend['date'].dt.year
df_trend['season'] = df_trend['month'].apply(
    lambda x: 'ë´„' if 3 <= x <= 5 else
             'ì—¬ë¦„' if 6 <= x <= 8 else
             'ê°€ì„' if 9 <= x <= 11 else 'ê²¨ìš¸'
)
```

```python
# %%
# ê³„ì ˆë³„ í‰ê·  ê²€ìƒ‰ëŸ‰
season_avg = df_trend.groupby('season')['ì„ í¬ë¦¼'].mean().sort_values(ascending=False)

print("ğŸ“Š ê³„ì ˆë³„ í‰ê·  ê²€ìƒ‰ëŸ‰:")
for season, value in season_avg.items():
    print(f"  {season}: {value:.2f}")

# %%
# ê²¨ìš¸/ì—¬ë¦„ ë¹„ìœ¨ ê³„ì‚°
summer_avg = season_avg['ì—¬ë¦„']
winter_avg = season_avg['ê²¨ìš¸']
ratio = winter_avg / summer_avg * 100

print(f"\nğŸ’¡ í•µì‹¬ ì§€í‘œ:")
print(f"  ì—¬ë¦„ í‰ê· : {summer_avg:.2f}")
print(f"  ê²¨ìš¸ í‰ê· : {winter_avg:.2f}")
print(f"  ê²¨ìš¸/ì—¬ë¦„ ë¹„ìœ¨: {ratio:.1f}%")
```

### ğŸ“ˆ ì‹œê°í™” 1: ê³„ì ˆë³„ ë¶„í¬

```python
# Plotly ì¸í„°ë™í‹°ë¸Œ ë°•ìŠ¤í”Œë¡¯ + ë°”ì´ì˜¬ë¦° í”Œë¡¯
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('ê³„ì ˆë³„ ë°•ìŠ¤í”Œë¡¯', 'ê³„ì ˆë³„ ë°”ì´ì˜¬ë¦° í”Œë¡¯'),
    specs=[[{"type": "box"}, {"type": "box"}]]
)

seasons_order = ['ë´„', 'ì—¬ë¦„', 'ê°€ì„', 'ê²¨ìš¸']
colors = {'ë´„': '#90EE90', 'ì—¬ë¦„': '#FFD700', 'ê°€ì„': '#FF6347', 'ê²¨ìš¸': '#87CEEB'}

# ë°•ìŠ¤í”Œë¡¯
for season in seasons_order:
    data = df_trend[df_trend['season'] == season]['ì„ í¬ë¦¼']
    fig.add_trace(
        go.Box(y=data, name=season, marker_color=colors[season], boxmean='sd'),
        row=1, col=1
    )

# ë°”ì´ì˜¬ë¦° í”Œë¡¯
for season in seasons_order:
    data = df_trend[df_trend['season'] == season]['ì„ í¬ë¦¼']
    fig.add_trace(
        go.Violin(y=data, name=season, marker_color=colors[season], showlegend=False),
        row=1, col=2
    )

fig.update_yaxes(title_text="ê²€ìƒ‰ëŸ‰ ì§€ìˆ˜", row=1, col=1)
fig.update_yaxes(title_text="ê²€ìƒ‰ëŸ‰ ì§€ìˆ˜", row=1, col=2)
fig.update_layout(
    title_text="ğŸ“Š ê³„ì ˆë³„ ì„ í¬ë¦¼ ê²€ìƒ‰ëŸ‰ ë¶„í¬ (ì¸í„°ë™í‹°ë¸Œ)",
    height=500,
    showlegend=True,
    hovermode='closest'
)
fig.show()
```

### ğŸ”¬ í†µê³„ì  ìœ ì˜ì„± ê²€ì •

```python

# ì—¬ë¦„ vs ê²¨ìš¸ t-test
summer_data = df_trend[df_trend['season']=='ì—¬ë¦„']['ì„ í¬ë¦¼']
winter_data = df_trend[df_trend['season']=='ê²¨ìš¸']['ì„ í¬ë¦¼']

t_stat, p_value = stats.ttest_ind(summer_data, winter_data)

print("ğŸ“ˆ t-ê²€ì • ê²°ê³¼:")
print(f"  t-í†µê³„ëŸ‰: {t_stat:.3f}")
print(f"  p-value: {p_value:.4f}")
print(f"  ììœ ë„: {len(summer_data) + len(winter_data) - 2}")

if p_value < 0.05:
    print(f"\n  âœ… ê²°ë¡ : ì—¬ë¦„ê³¼ ê²¨ìš¸ì˜ ê²€ìƒ‰ëŸ‰ ì°¨ì´ëŠ” í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•©ë‹ˆë‹¤ (p<0.05)")
else:
    print(f"\n  âš ï¸ ê²°ë¡ : ì—¬ë¦„ê³¼ ê²¨ìš¸ì˜ ê²€ìƒ‰ëŸ‰ ì°¨ì´ëŠ” í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤ (pâ‰¥0.05)")
    
# íš¨ê³¼ í¬ê¸° (Cohen's d)
pooled_std = np.sqrt(((len(summer_data)-1)*summer_data.std()**2 + 
                      (len(winter_data)-1)*winter_data.std()**2) / 
                     (len(summer_data) + len(winter_data) - 2))
cohens_d = (summer_data.mean() - winter_data.mean()) / pooled_std

print(f"\nğŸ“ íš¨ê³¼ í¬ê¸° (Cohen's d): {cohens_d:.3f}")
if abs(cohens_d) < 0.5:
    print("  â†’ ì‘ì€ íš¨ê³¼")
elif abs(cohens_d) < 0.8:
    print("  â†’ ì¤‘ê°„ íš¨ê³¼")
else:
    print("  â†’ í° íš¨ê³¼")
```

### ğŸ“… ì—°ë„ë³„ ê²¨ìš¸ íŠ¸ë Œë“œ

```python
# ì—°ë„ë³„ ê²¨ìš¸ ê²€ìƒ‰ëŸ‰
df_winter = df_trend[df_trend['season']=='ê²¨ìš¸'].copy()
yearly_winter = df_winter.groupby('year')['ì„ í¬ë¦¼'].mean()

print("ğŸ“… ì—°ë„ë³„ ê²¨ìš¸ í‰ê·  ê²€ìƒ‰ëŸ‰:")
for year, value in yearly_winter.items():
    print(f"  {year}ë…„: {value:.2f}")
```

### ğŸ¯ ì¸ì‚¬ì´íŠ¸ ìš”ì•½

```python
# ì¦ê°€ìœ¨ ê³„ì‚°
growth_rates = yearly_winter.pct_change() * 100
print(f"\nğŸ“ˆ ì „ë…„ ëŒ€ë¹„ ì¦ê°€ìœ¨:")
for year, rate in growth_rates.dropna().items():
    print(f"  {year}ë…„: {rate:+.1f}%")
```

```python
# ì—¬ë¦„/ê²¨ìš¸ ë¹„êµ + ì—°ë„ë³„ íŠ¸ë Œë“œ (ì„œë¸Œí”Œë¡¯)
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=('ì—°ë„ë³„ ê²¨ìš¸ vs ì—¬ë¦„ ê²€ìƒ‰ëŸ‰ ì¶”ì´', 'ì—°ë„ë³„ ê²¨ìš¸ ê²€ìƒ‰ëŸ‰ ì„±ì¥ìœ¨'),
    specs=[[{"secondary_y": False}], [{"secondary_y": False}]],
    vertical_spacing=0.12
)

# 1ë²ˆ ê·¸ë˜í”„: ì—°ë„ë³„ ê²¨ìš¸/ì—¬ë¦„ ë¹„êµ
yearly_summer = df_trend[df_trend['season']=='ì—¬ë¦„'].groupby('year')['ì„ í¬ë¦¼'].mean()
yearly_winter = df_trend[df_trend['season']=='ê²¨ìš¸'].groupby('year')['ì„ í¬ë¦¼'].mean()

fig.add_trace(
    go.Scatter(
        x=yearly_winter.index.astype(str), 
        y=yearly_winter.values,
        mode='lines+markers',
        name='ê²¨ìš¸',
        line=dict(color='#87CEEB', width=3),
        marker=dict(size=10, symbol='circle')
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=yearly_summer.index.astype(str), 
        y=yearly_summer.values,
        mode='lines+markers',
        name='ì—¬ë¦„',
        line=dict(color='#FFD700', width=3),
        marker=dict(size=10, symbol='diamond')
    ),
    row=1, col=1
)

# 2ë²ˆ ê·¸ë˜í”„: ì„±ì¥ìœ¨
growth_rates = yearly_winter.pct_change() * 100
colors_growth = ['green' if x > 0 else 'red' for x in growth_rates.dropna()]

fig.add_trace(
    go.Bar(
        x=growth_rates.dropna().index.astype(str),
        y=growth_rates.dropna().values,
        marker_color=colors_growth,
        name='ì„±ì¥ìœ¨',
        showlegend=False
    ),
    row=2, col=1
)

fig.add_hline(y=0, line_dash="dash", line_color="black", row=2, col=1)

fig.update_xaxes(title_text="ì—°ë„", row=1, col=1)
fig.update_xaxes(title_text="ì—°ë„", row=2, col=1)
fig.update_yaxes(title_text="ê²€ìƒ‰ëŸ‰ ì§€ìˆ˜", row=1, col=1)
fig.update_yaxes(title_text="ì „ë…„ ëŒ€ë¹„ ì¦ê°€ìœ¨ (%)", row=2, col=1)

fig.update_layout(
    title_text="ğŸ“ˆ ê²¨ìš¸ ì„ í¬ë¦¼ ê²€ìƒ‰ëŸ‰ ì—°ë„ë³„ ë³€í™” ë¶„ì„",
    height=700,
    hovermode='x unified'
)
fig.show()
```

```python
print("=" * 60)
print("ğŸ’¡ Q1 ì¸ì‚¬ì´íŠ¸")
print("=" * 60)

print(f"\n1ï¸âƒ£ ì‹œì¥ ê·œëª¨:")
print(f"   ê²¨ìš¸ ì‹œì¥ì€ ì—¬ë¦„ ëŒ€ë¹„ {ratio:.1f}%ì˜ ìˆ˜ìš”ë¥¼ ë³´ì„")

if ratio > 30:
    print(f"   âœ… íˆ¬ì ê°€ì¹˜ ì¶©ë¶„ (30% ì´ìƒ)")
elif ratio > 20:
    print(f"   âš ï¸ ì¤‘ê°„ ìˆ˜ì¤€ ì‹œì¥ (20-30%)")
else:
    print(f"   âŒ ì†Œê·œëª¨ ì‹œì¥ (20% ë¯¸ë§Œ)")

print(f"\n2ï¸âƒ£ í†µê³„ì  ìœ ì˜ì„±:")
print(f"   p-value = {p_value:.4f} â†’ {'ìœ ì˜ë¯¸' if p_value < 0.05 else 'ìœ ì˜ë¯¸í•˜ì§€ ì•ŠìŒ'}")

print(f"\n3ï¸âƒ£ ì„±ì¥ ì¶”ì„¸:")
if len(growth_rates.dropna()) > 0:
    avg_growth = growth_rates.dropna().mean()
    print(f"   ì—°í‰ê·  ì¦ê°€ìœ¨: {avg_growth:+.1f}%")
    if avg_growth > 0:
        print(f"   âœ… ì‹œì¥ ì„±ì¥ ì¤‘")
    else:
        print(f"   âš ï¸ ì‹œì¥ ì •ì²´/ê°ì†Œ")
```
